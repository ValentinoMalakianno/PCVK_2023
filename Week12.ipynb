**Membuat Fungsi untuk Memudahkan Konversi Warna Citra**

```python
def bgr_to_rgb(images):
  return cv2.cvtColor(images, cv2.COLOR_BGR2RGB)

def gray_to_rgb(images):
  return cv2.cvtColor(images, cv2.COLOR_GRAY2RGB)

def bgr_to_gray(images):
  return cv2.cvtColor(images, cv2.COLOR_BGR2GRAY)
```

**2. Face Detection untuk Beberapa Citra**

```python
# Mendefinisikan Cascade Classifier
cascade_wajah = cv2.CascadeClassifier('/content/drive/MyDrive/Face Detection/haarcascade_frontalface_alt.xml')

# Jokowi

jokowi_bgr = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/jokowi.jpg')
jokowi = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/jokowi.jpg')

# Mendeteksi wajah
roi_wajah = cascade_wajah.detectMultiScale(jokowi_bgr)

# Menggambar kotak hijau di sekitar wajah pada gambar BGR
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(jokowi_bgr, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Mengonversi gambar BGR ke RGB menggunakan fungsi bgr_to_rgb
jokowi_rgb = bgr_to_rgb(jokowi_bgr)

# Menampilkan gambar dengan kotak hijau di sekitar wajah
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(bgr_to_rgb(jokowi))
plt.axis('off')
plt.subplot(122), plt.imshow(jokowi_rgb)
plt.axis('off')
plt.show()


# RA. Kartini

kartini_bgr = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/kartini.jpg')
kartini = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/kartini.jpg')

# Mendeteksi wajah
roi_wajah = cascade_wajah.detectMultiScale(kartini_bgr)

# Menggambar kotak hijau di sekitar wajah pada gambar BGR
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(kartini_bgr, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Mengonversi gambar BGR ke RGB menggunakan fungsi bgr_to_rgb
kartini_rgb = bgr_to_rgb(kartini_bgr)

# Menampilkan gambar dengan kotak hijau di sekitar wajah
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(bgr_to_rgb(kartini))
plt.axis('off')
plt.subplot(122), plt.imshow(kartini_rgb)
plt.axis('off')
plt.show()

# Kucing

kucing_bgr = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/kucing.jpg',0)
kucing = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/kucing.jpg',0)

# Mendeteksi wajah
roi_wajah = cascade_wajah.detectMultiScale(kucing_bgr)

# Menggambar kotak hijau di sekitar wajah pada gambar BGR
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(kucing_bgr, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Mengonversi gambar BGR ke RGB menggunakan fungsi bgr_to_rgb
kucing_rgb = bgr_to_rgb(kucing_bgr)

# Menampilkan gambar dengan kotak hijau di sekitar wajah
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(kucing, cmap='gray')
plt.axis('off')
plt.subplot(122), plt.imshow(kucing_bgr, cmap='gray')
plt.axis('off')
plt.show()

# Perempuan bermasker

mask_gray = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/mask.png',0)

# Mendeteksi wajah
roi_wajah = cascade_wajah.detectMultiScale(mask_gray)

# Mengonversi gambar grayscale ke RGB untuk menyimpan warna hijau
mask_gray_rgb = gray_to_rgb(mask_gray)

# Menggambar kotak hijau di sekitar wajah pada gambar BGR
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(mask_gray_rgb, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Menampilkan gambar dengan kotak hijau di sekitar wajah
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(mask_gray, cmap='gray')
plt.axis('off')
plt.subplot(122), plt.imshow(mask_gray_rgb)
plt.axis('off')
plt.show()

# Jordan

jordan_gray = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/mjordan.jpg',0)

# Mendeteksi wajah
roi_wajah = cascade_wajah.detectMultiScale(jordan_gray)

# Mengonversi gambar grayscale ke RGB untuk menyimpan warna hijau
jordan_gray_rgb = gray_to_rgb(jordan_gray)

# Menggambar kotak hijau di sekitar wajah pada gambar BGR
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(jordan_gray_rgb, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Menampilkan gambar dengan kotak hijau di sekitar wajah
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(jordan_gray, cmap='gray')
plt.axis('off')
plt.subplot(122), plt.imshow(jordan_gray_rgb)
plt.axis('off')
plt.show()

# Solvayconf

solvay_gray = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/solvayconf.jpg', 0)

# Mendeteksi wajah
roi_wajah = cascade_wajah.detectMultiScale(solvay_gray)

# Mengonversi gambar grayscale ke RGB untuk menyimpan warna hijau
solvay_gray_rgb = gray_to_rgb(solvay_gray)

# Menggambar kotak hijau di sekitar wajah pada gambar grayscale
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(solvay_gray_rgb, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Menampilkan kedua gambar
plt.figure(figsize=(8, 8))

plt.subplot(121), plt.imshow(solvay_gray, cmap='gray')
plt.axis('off')

plt.subplot(122), plt.imshow(solvay_gray_rgb)
plt.axis('off')

plt.show()
```

```python
# Kucing

# Mendefinisikan Cascade Classifier
cascade_wajah = cv2.CascadeClassifier('/content/drive/MyDrive/Face Detection/haarcascade_frontalcatface.xml')

kucing = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/kucing.jpg')
kucing_bgr = cv2.imread('/content/drive/MyDrive/PCVK_2023/Images/facedet/kucing.jpg')
# Mendeteksi wajah
roi_wajah = cascade_wajah.detectMultiScale(kucing_bgr)

# Mengonversi gambar grayscale ke RGB untuk menyimpan warna hijau
# kucing_gray_rgb = gray_to_rgb(kucing_gray)

# Menggambar kotak hijau di sekitar wajah pada gambar BGR
for (x, y, w, h) in roi_wajah:
    cv2.rectangle(kucing_bgr, (x, y), (x+w, y+h), (0, 255, 0), 3)

# Mengonversi gambar BGR ke RGB menggunakan fungsi bgr_to_rgb
kucing_rgb = bgr_to_rgb(kucing_bgr)

# Menampilkan gambar dengan kotak hijau di sekitar wajah
plt.figure(figsize=(8, 8))
plt.subplot(121), plt.imshow(bgr_to_rgb(kucing))
plt.axis('off')
plt.subplot(122), plt.imshow(kucing_rgb)
plt.axis('off')
plt.show()
```
